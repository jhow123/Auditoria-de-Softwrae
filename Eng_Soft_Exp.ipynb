```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Número de amostras a serem geradas
n = 350

data = {
    'Interface': np.random.randint(1, 6, n),
    'Desempenho': np.random.randint(1, 6, n),
    'Funcionalidades': np.random.randint(1, 6, n)
}

# Criar o DataFrame
df = pd.DataFrame(data)

# Calcular as estatísticas descritivas
summary = df.describe()

# Gráfico de Barras para Médias por Categoria
plt.figure(figsize=(10, 6))
mean_ratings = df.mean()
mean_ratings.plot(kind='bar', color='skyblue')
plt.title('Médias de Avaliação por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Média de Avaliação')
plt.xticks(rotation=0)
plt.show()

# Gráfico de Dispersão com Barras de Erro para Variabilidade
plt.figure(figsize=(10, 6))
sns.barplot(data=df, capsize=0.1)
plt.title('Variabilidade das Avaliações por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Avaliação')
plt.show()
```
